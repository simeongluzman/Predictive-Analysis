```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Load the original data
data = pd.read_csv('your_data.csv')

# Define the columns to be encoded
cat_cols = ['learning_goal', 'topic']
label_cols = ['selected_track']

# Define the encoding schemes for each column
cat_encoder = OneHotEncoder()
label_encoder = LabelEncoder()

# Define the column transformer
preprocessor = ColumnTransformer(transformers=[
    ('cat', cat_encoder, cat_cols),
    ('label', label_encoder, label_cols)
])

# Fit and transform the data
X_encoded = preprocessor.fit_transform(data)

# Convert the encoded data back to a DataFrame
cols = cat_encoder.get_feature_names(cat_cols).tolist() + label_cols
X_encoded_df = pd.DataFrame(X_encoded.toarray(), columns=cols)

# View the encoded data
print(X_encoded_df.head())
```
